```python
def my_square(num):
    """
    Instrukcja funkcji - robi kwadrat w Jupiter shift+Tab po wpisaniu funkcji
    """
    return num**2 # return zwraca wynik do dalszej obróbki nie tylko go drukuje

out = my_square(3)
print(out)
```

```python
seq = [1,2,3,4,5]
x = map(my_square, seq) # map wykonuje daną funkcje/wyrażenie dla wszystkich liczb w danej sekwencji
list(x)
```

```python
# funkcja filter - jak map ale oddaje liczby spełniające warunek

x = filter(lambda num: num%2 == 0, seq)
list(x)
```

```python
# lambda - niezdefiniowana funkcja
lambda var: var**2 #robi to samo co funcja my_square

list(map(lambda num: num**2,seq))
```

```python
import seaborn as sns 
%matplotlib inline
sns.set()

# w seaborn dostępne są wybrane zbiory danych
tips = sns.load_dataset("tips")
tips.head()
```

```python
# pokazuje rozkład pojedyńczej zmiennej # jest różnica między displot a distplot
sns.distplot(tips['total_bill']) 
```

```python
# kde = True - rozkład, Flase - nie ma rozkładu # bins - ilość kolumn
sns.distplot(tips['total_bill'],kde=False, bins = 20) 
```

```python
# rozkład 2 wartości i wykres ich zależności / korelację 
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg') #, kind ='hex','kde'
```

```python
# wykonuje jointplot dla całego zestawu danych
sns.pairplot(tips,hue='sex') #,pallete  # hue wyróżnia katergorię danych - dzieli dane w zależności od płci 
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
# z wykresu wynika że mężczyźni płącą więcej w restaturacjach
```

```python
# countplot wypisuje ilość próbek w danej kategorii - w danych jest więcej mężczyzn
sns.countplot(x='sex',data=tips)
```

```python
# wykres rachunków od dni tygodnia
sns.boxplot(x='day',y='total_bill',data=tips) 
# pudełka pokazują dwie ćwiartki rozkładu normalnego, czarna kresta wewnątrz to mediana, linie poza polazują skrajne
# ćwiartki rozkładu a kropki poza są punktami ekstremalnymi wychodzącymi poza rozkad
```

```python
import chart_studio.plotly as py 
import cufflinks as cf 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
%matplotlib inline
init_notebook_mode(connected=True) 
```

```python
import plotly.graph_objs as go
```

```python
# dict - tworzy słownik z listy, choropleth - typ map w plotly
data = dict(type = 'choropleth',
            locations = ['AZ','CA','NY'], # określa wybrane lokacje 
            locationmode = 'USA-states',  # określa położenie w swiecie
            colorscale= 'Portland', # określa schemat kolorów
            text= ['Arizona','text2','New York'],
            z=[1.0,2.0,3.0],
            colorbar = {'title':'Colorbar'})
```

```python
layout = dict(geo = {'scope':'usa'})
```

```python
# do stworzenia mapy potrzebujemy 2 obiektów data i laytout 
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```

```python
# zbiór danych o eksporcie z każdego stanu
df = pd.read_csv('Refactored_Py_DS_ML_Bootcamp-master/09-Geographical-Plotting/2011_US_AGRI_Exports')
df.head()
```

```python
data = dict(type = 'choropleth',
            colorscale = 'ylorbr',
            locations = df['code'],
            locationmode = 'USA-states', 
            z = df['total exports'],
            text = df['text'],
            marker = dict(line = dict(color = 'rgb(12,12,12)', width=2)),
            colorbar = {'title': 'Milions USD'})
```

```python
layout = dict(title = '2011 US Exports by State',
              geo = dict(scope='usa', showlakes=True, lakecolor='rgb(85,173,240)'))
```

```python
choromap2 = go.Figure(data = [data], layout=layout)
iplot(choromap2)
```
